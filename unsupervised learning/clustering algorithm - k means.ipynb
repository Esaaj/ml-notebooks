### ***Cluster Based Algorithm***
Cluster based algorithm is a type of unsupervised machine learning technique used to group similar data points into clusters based on their features. 

These algorithms analyze the inherent structure of the data and identify patterns without prior knowledge of the labels or categories. 

Common examples of cluster-based algorithms include K-means, hierarchical clustering, and DBSCAN. They are widely used in various applications such as customer segmentation, image analysis, and anomaly detection.


***K-Means Clustering***
K-means is a popular cluster-based algorithm that partitions data into K distinct clusters based on feature similarity. 

The algorithm works as follows:
1. Initialization: Select K initial centroids randomly from the dataset.
2. Assignment Step: Assign each data point to the nearest centroid, forming K clusters.
3. Update Step: Calculate the new centroids by taking the mean of all data points assigned to each cluster.
4. Repeat: Repeat the assignment and update steps until the centroids no longer change significantly or a maximum number of iterations is reached.

K-means is widely used due to its simplicity and efficiency, but it requires the user to specify the number of clusters (K) in advance and can be sensitive to the initial placement of centroids.


*** key terms ***

| Term                 | Meaning                                                                |
| -------------------- | ---------------------------------------------------------------------- |
| **K**                | Number of clusters                                                     |
| **Centroid**         | Center of a cluster                                                    |
| **Inertia**          | Measure of cluster compactness (sum of squared distances to centroids) |
| **Elbow Method**     | Technique to find optimal K                                            |
| **Silhouette Score** | Measures how well data points fit their assigned cluster               |


***Elbow method*** is a technique used to determine the optimal number of clusters (K) in K-means clustering. 

It involves plotting the explained variance (or inertia) against the number of clusters and identifying the "elbow" point where the rate of decrease sharply changes. 

This point indicates a balance between cluster compactness and simplicity, helping to choose an appropriate K value for the clustering task.

***Silhouette Score***

It is a metric used to evaluate the quality of clustering in unsupervised learning. It measures how similar an object is to its own cluster compared to other clusters.
  
The silhouette score ranges from -1 to 1, where a score close to 1 indicates that the data point is well clustered, a score of 0 indicates that the data point is on or very close to the decision boundary between two neighboring clusters, and a score close to -1 indicates that the data point may have been assigned to the wrong cluster. The silhouette score can be calculated for individual data points as well as for the entire clustering solution by averaging the scores of all data points.

````python
# Implementation of K-means clustering algorithm in Python using scikit-learn:

import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.datasets import make_blobs

# Generate synthetic data
X, y_true = make_blobs(n_samples=300, centers=4, cluster_std=0.60, random_state=0)
plt.scatter(X[:, 0], X[:, 1], s=30)
plt.title("Synthetic Data")
plt.show()

# Apply K-means clustering
kmeans = KMeans(n_clusters=4)
kmeans.fit(X)
y_kmeans = kmeans.predict(X)

# Plot the clustered data
plt.scatter(X[:, 0], X[:, 1], c=y_kmeans, s=30, cmap='viridis')
centers = kmeans.cluster_centers_
plt.scatter(centers[:, 0], centers[:, 1], c='red', s=200, alpha=0.75, marker='X')
plt.title("K-means Clustering")
plt.show()  

# Evaluate clustering using inertia
print("Inertia:", kmeans.inertia_)  
# Inertia is the sum of squared distances of samples to their closest cluster center.
# Lower inertia indicates better clustering.
# You can also use the elbow method to determine the optimal number of clusters.
# Elbow Method
inertia = []
K = range(1, 10)
for k in K:
    kmeans = KMeans(n_clusters=k)
    kmeans.fit(X)
    inertia.append(kmeans.inertia_)
plt.plot(K, inertia, 'bx-')
plt.xlabel('Number of clusters K')
plt.ylabel('Inertia')
plt.title('Elbow Method For Optimal K')
plt.show()

# Calculate silhouette score
from sklearn.metrics import silhouette_score
score = silhouette_score(X, y_kmeans)
print("Silhouette Score:", score)
# Silhouette score measures how similar an object is to its own cluster compared to other clusters.
# A higher silhouette score indicates better-defined clusters.
````

**Order of Implementation for K-Means Clustering** 👇

1. **Import necessary libraries**
   (`numpy`, `pandas`, `matplotlib`, `sklearn.cluster.KMeans`, etc.)

2. **Load the dataset**
   (CSV file, generated data, or from sklearn datasets)

3. **Preprocess the data**

   * Handle missing values
   * Scale or normalize features if needed (`StandardScaler` or `MinMaxScaler`)

4. **Select number of clusters (K)**

   * Use prior knowledge or find it using **Elbow Method** or **Silhouette Score**

5. **Initialize and fit the K-Means model**

   ```python
   kmeans = KMeans(n_clusters=K, init='k-means++', random_state=42)
   kmeans.fit(X)
   ```

6. **Predict cluster labels**

   ```python
   labels = kmeans.predict(X)
   ```

7. **Analyze cluster centers (centroids)**

   ```python
   centers = kmeans.cluster_centers_
   ```

8. **Visualize the clusters**

   * 2D/3D scatter plot of clusters
   * Highlight centroids in red or larger points

9. **Evaluate performance** *(optional)*

   * **Inertia (SSE)** → `kmeans.inertia_`
   * **Silhouette Score** → `silhouette_score(X, labels)`

10. **Tune K (optional)**

    * Try different K values and compare results.

### ***Ploynomial Regression*** ###

It is a type of Linear Regression where the relationship between the independent variable (X) and dependent variable (y) is non-linear ‚Äî but we model it using a polynomial equation.

```math
    y = b0 ‚Äã+ b1‚Äãx + b2‚Äãx2 + b3‚Äãx3 + ‚ãØ + bn‚Äãxn
```

So instead of a straight line, the regression curve can bend and follow curves in data.

Addition of non-linear features(x^2, x^3, x^4) which makes linear regression model non-linear is called ploynomial regression.

***Bias*** is nothing but training error on training data. calculated as the difference between average prediction of our model and the correct value which we are trying to predict.

***Variance*** is the variability of model prediction for a given data point or a value which tells us spread of our data. 

***Bestfit*** means model has low variance and bias. 
    
***Underfit*** means model is too simple to learn the underlying pattern of data. It has high bias and low variance.

***overfit*** means model is too complex and learns the noise of data. It has low bias and high variance.

We need to find the right polynomial degree (n) that fits the data just right ‚Äî
‚úÖ not underfitting (too simple)
‚ùå not overfitting (too complex)

Realiable Method to find the best polynomial degree.

You train and evaluate models for different degrees (say 1‚Äì6) and compare their cross-validation scores.

For each degree of the polynomial, you calculate the R¬≤ score using cross-validation.

Then pick the degree with the highest R¬≤ score (but before it starts dropping).

### ***k - nearest neighbours (KNN)*** ###
KNN is a supervised learning algorithm used for classification and regression.

The idea: ‚ÄúA data point is classified based on how its neighbors are classified.‚Äù

It uses distance metrics (like Euclidean distance) to find the K nearest data points, then predicts the output based on the majority (for classification) or average (for regression).

üß© How it works

Choose the number of neighbors, K.

Calculate the distance between the new data point and all other points in the dataset.

Select the K closest points.

For classification ‚Üí take the majority class among neighbors.
For regression ‚Üí take the average of neighbors‚Äô values.

Return the result.

***How to find the best k value in KNN algorithm***

Predict the output based on the majority (for classification) or average (for regression) using k values from 1 to 20 and plot the accuracy for each k value.

using the elbow method to find the best k value.

```python
# implementation of KNN algorithm in Python

from sklearn.neighbors import KNeighborsClassifier

# Create a KNN model
model = KNeighborsClassifier(n_neighbors=3)

# other parameters: weights='uniform', algorithm='auto', leaf_size=30, p=2, metric='minkowski', metric_params=None, n_jobs=None
# where
# n_neighbors: Number of neighbors to use (default is 5)
# weights: Weight function used in prediction (default is 'uniform')
# algorithm: Algorithm used to compute the nearest neighbors (default is 'auto')
# leaf_size: Leaf size passed to BallTree or KDTree (default is 30)
# p: Power parameter for the Minkowski metric (default is 2)
# metric: The distance metric to use (default is 'minkowski')
# metric_params: Additional keyword arguments for the metric function (default is None)
# n_jobs: Number of parallel jobs to run for neighbors search (default is None)
```
